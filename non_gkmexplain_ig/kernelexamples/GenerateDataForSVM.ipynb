```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random
```

```python
def isFarEnoughFromLine(p1, distance):
    # Assume line is x = y
    x = p1[0]
    y = p1[1]
    return (y >= (x + distance)) or (y <= (x - distance)) 
```

```python
def lineLabel(p1):
    x = p1[0]
    y = p1[1]
    if x > y:
        return 1
    else:
        return 0
```

```python
def generatePoints(size, distance, labellingFunction, distanceCheck):    
    points = np.zeros((size,2))
    labels = np.zeros(size, dtype=np.int)
    i=0
    accepted = set()
    while i < size:
        x = random.randint(-50, 50)
        y = random.randint(-50, 50)
        point = (x,y)
        if not (point in accepted) and distanceCheck(point, distance):
            accepted.add(point)
            points[i, 0] = x
            points[i, 1] = y
            labels[i] = labellingFunction(point)
            i = i+1
    return (points, labels)
```

```python
(points, labels) = generatePoints(1000, 1, lineLabel, isFarEnoughFromLine)
```

```python
#print(points)
#print(labels)
```

```python
plt.scatter(points[:,0], points[:,1], c=labels[:])
plt.show()
```

```python
def isFarEnoughFromCircle(p1, distance):
    x = float(p1[0])
    y = float(p1[1])
    d = float(distance)
    # Assume circle of radius 30
    pointRadius = math.sqrt(x**2 + y**2)
    return (pointRadius >= 30+d) or (pointRadius <= 30-d)
    
```

```python
def circleLabel(p1):
    x = float(p1[0])
    y = float(p1[1])
    #Assume circle of radius 30
    pointRadius = math.sqrt(x**2 + y**2)
    if pointRadius > 30:
        return 0
    else:
        return 1
```

```python
(points, labels) = generatePoints(1000, 1, circleLabel, isFarEnoughFromCircle)
```

```python
plt.scatter(points[:,0], points[:,1], c=labels[:])
plt.show()
```

```python
def isFarEnoughFromParabola(p1, distance):
    x = float(p1[0])
    y = float(p1[1])
    d = float(distance)
    # Assume parabola is y = 0.25*x**2 - 40
    # Distance calculation is inaccurate, but truly parallel parabolas separated by fixed distance 
    # are hard
    return (y >= 0.25*x**2 - 40 + d) or (y <= 0.25*x**2 - 40 - d) 
```

```python
def parabolaLabel(p1):
    x = float(p1[0])
    y = float(p1[1])
    #Assume parabola is y = 0.25*x**2 - 40
    if y > 0.25*x**2 - 40:
        return 1
    else:
        return 0
```

```python
(points, labels) = generatePoints(1000, 1, parabolaLabel, isFarEnoughFromParabola)
```

```python
plt.scatter(points[:,0], points[:,1], c=labels[:])
plt.show()
```

```python
def upperRightQuadrantLabel(p1):
    x = float(p1[0])
    y = float(p1[1])
    if x > 0 and y > 0:
        return 1
    else:
        return 0
```

```python
def isFarEnoughFromQuadrantBoundary(p1, distance):
    x = float(p1[0])
    y = float(p1[1])
    d = float(distance)
    if (abs(x) < distance) and (abs(y) < distance):
        return False
    if (x>0) and (abs(y) < distance):
        return False
    if (y>0) and (abs(x) < distance):
        return False
    return True
```

```python
(points, labels) = generatePoints(1000, 1, upperRightQuadrantLabel, isFarEnoughFromQuadrantBoundary)
```

```python
plt.scatter(points[:,0], points[:,1], c=labels[:])
plt.show()
```
