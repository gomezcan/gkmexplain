```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np 
```

```python
ig10x, ig10y = np.loadtxt('ig10.txt', delimiter=',', unpack=True)
shap10x, shap10y = np.loadtxt('shap10.txt', delimiter=',', unpack=True)
shap50x, shap50y = np.loadtxt('shap50.txt', delimiter=',', unpack=True)
lime50x, lime50y = np.loadtxt('lime50.txt', delimiter=',', unpack=True)
```

```python
ig10ymean = np.mean(ig10y)
shap10ymean = np.mean(shap10y)
shap50ymean = np.mean(shap50y)
lime50ymean = np.mean(lime50y)
```

```python
plt.ylim((0,1))
plt.xlim((0,15))
plt.title("Runtime vs Accuracy")
plt.xlabel("Runtime in seconds")
plt.ylabel("SVM accuracy using 10 most important features")
ig10 = plt.scatter(ig10x,ig10y,color="blue")
ig10mean = plt.plot([0, 15], [ig10ymean, ig10ymean], color="blue", label='IG 10 average') 
shap10 = plt.scatter(shap10x, shap10y,color="red")
shap10mean = plt.plot([0, 15], [shap10ymean, shap10ymean], color="red", label='SHAP 10 average' ) 
shap50 = plt.scatter(shap50x, shap50y,color="purple")
shap50mean = plt.plot([0, 15], [shap50ymean, shap50ymean], color="purple", label='SHAP 50 average') 
lime50 = plt.scatter(lime50x, lime50y,color="green")
lime50mean = plt.plot([0, 15], [lime50ymean, lime50ymean], color="green", label='LIME 50 average') 
plt.legend((ig10, shap10, shap50, lime50),
           ('IG 10 Points', 'SHAP 10 Samples', 'SHAP 50 Samples', 'LIME 50 Samples'), 
           scatterpoints=1,
           loc='lower right',
           ncol=2,
           fontsize=9
          )
plt.savefig("ImpScoringAlgWithMean.png")
plt.show()
```
